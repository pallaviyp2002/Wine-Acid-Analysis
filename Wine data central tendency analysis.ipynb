```python
plt.figure(figsize=(10,5))
sns.histplot(data=df,x="volatile acidity",color = "blue",edgecolor = "pink",bins = 5)
plt.title("Histogram of volatile acidity")
plt.xlabel("volatile acidity")
plt.ylabel("counts")
plt.show()
```

# this histogram is positively skewed (asymmetric), hence we need to plot normal distplot

```python
plt.figure(figsize=(10,5))
sns.distplot(df["volatile acidity"],color = "orange")
plt.title("Normal distribution of volatile acidity")
plt.xlabel("volatile acidity")
plt.ylabel("counts")
plt.show()
```

#  Normal distribution of volatile acidity
It has mean and median are same
It has only mode
It is positively skewed curve

```python
df["volatile acidity"].skew()
```

# It is more than one hence it is positively skewed

```python
df["volatile acidity"].mean()
```

```python
df["volatile acidity"].median()
```

# by the observation we say that both mean and median are equal

```python
plt.figure(figsize = (10,5))
sns.histplot(data=df,x="volatile acidity",color="red",edgecolor="blue",bins=5)
plt.title("Histogram of volatile acidity")
plt.xlabel("volatile acidity")
plt.ylabel("count")
plt.vlines(df["volatile acidity"].mean(),ymin=0,ymax=4000,color="black",label="mean")
plt.vlines(df["volatile acidity"].median(),ymin=0,ymax=4000,color="orange",label="median")
plt.show()
```
